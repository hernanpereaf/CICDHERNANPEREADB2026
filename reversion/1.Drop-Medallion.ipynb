```sql
%python
dbutils.widgets.removeAll()
```

```sql
create widget text storageName default "adlsproyecto2026";
create widget text catalogo default "ventas";
```

```sql
%python
storageName = dbutils.widgets.get("storageName")
catalogo = dbutils.widgets.get("catalogo")
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.sales;
DROP TABLE IF EXISTS ${catalogo}.bronze.stores;
DROP TABLE IF EXISTS ${catalogo}.bronze.brands;
```

```sql
DROP SCHEMA IF EXISTS  ${catalogo}.bronze;
```

```sql

%python
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/sales",True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/stores",True)
dbutils.fs.rm(f"abfss://bronze@{storageName}.dfs.core.windows.net/brands",True)
```

### Eliminacion tablas Silver

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.silver.sales_stores_transformed;
```

```sql
%python
dbutils.fs.rm(f"abfss://silver@{storageName}.dfs.core.windows.net/sales_stores_transformed",True)
```

### Eliminacion tablas Golden

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.golden.golden_sales_partitioned;
```

```sql
%python
# REMOVE DATA (Golden)
dbutils.fs.rm(f"abfss://golden@{storageName}.dfs.core.windows.net/golden_sales_partitioned",True)
```
